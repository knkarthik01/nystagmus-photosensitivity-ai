Create synthetic dataset for training our model.

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 2: Generate synthetic data

# Simulate environmental brightness and eye movement variance
data = pd.DataFrame({
    'brightness_level': np.random.uniform(100, 1000, 500),
    'eye_movement_variance': np.random.uniform(0, 10, 500),
    'photosensitivity_risk': np.random.choice([0, 1], 500)
})

data.head()
```

```python
# Step 3: Visualize synthetic data
plt.scatter(data['brightness_level'], data['eye_movement_variance'], c=data['photosensitivity_risk'])
plt.xlabel('Brightness Level')
plt.ylabel('Eye Movement Variance')
plt.title('Synthetic Data for Photosensitivity Risk')
plt.show()
```

```python
# Step 4: Save the preprocessed data
data.to_csv('preprocessed_data.csv', index=False)
print('âœ… Data saved as preprocessed_data.csv')
```
